```python
# simulate the low pass filter

# import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the function
def low_pass_filter(x, y, dt, RC):
    """
    x: input signal
    y: output signal
    dt: time step
    RC: time constant
    """
    alpha = dt/(RC+dt)
    for i in range(1, len(x)):
        y[i] = alpha*x[i] + (1-alpha)*y[i-1]
    return y
```

```python
# define the time constant
RC = .01
# define the time step
dt = 0.01

# generate the input signal
t = np.arange(0, 10, 0.01)
# x1 is the first input signal that needs to be filtered
x1 = np.sin(t) + np.random.randn(len(t))*0.1
# x2 is the second input signal that needs to be filtered
x2 = np.sin(t) + np.sin(100*t)*0.1
# x3 is the third input signal that needs to be filtered
x3 = np.sin(t) + np.sin(100*t)*0.1 + np.random.randn(len(t))*0.1


```

```python
# filter the input signal
y1 = np.zeros(len(x1))
y2 = np.zeros(len(x2))
y3 = np.zeros(len(x3))

y1 = low_pass_filter(x1, y1, 0.01, RC)
y2 = low_pass_filter(x2, y2, 0.01, RC)
y3 = low_pass_filter(x3, y3, 0.01, RC)

# plot the input and output signal
plt.figure(figsize=(10, 10))
plt.subplot(3, 1, 1)
plt.plot(t, x1, label='input')
plt.plot(t, y1, label='output')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, x2, label='input')
plt.plot(t, y2, label='output')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, x3, label='input')
plt.plot(t, y3, label='output')
plt.legend()
plt.show()
```
